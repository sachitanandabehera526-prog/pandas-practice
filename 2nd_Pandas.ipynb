# Series

```python
import pandas as pd
x=[1,2,3,4,5]
var = pd.Series(x,index=['a','b','c','d','e'],dtype="float")
print(var)
```

# Arithmetic Operations

```python
import pandas as pd

var = pd.DataFrame({"A":[1,2,3,4], "B":[5,6,7,8]})
var
```

```python
var["C"] = var["A"]+var["B"]
var
```

```python
var["C"] = var["A"]-var["B"]
var
```

```python
var["C"] = var["A"]*var["B"]
var
```

```python
var["C"] = var["A"]/var["B"]
var
```
